```python
# Librerias
import numpy as np # generar arreglos
import pandas as pd # cargar conjuntos de datos externos
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
# Cargar datos
np.random.seed(0)
horas_estudio = np.random.rand(100, 1) * 10
calificaciones = 2 * horas_estudio + np.random.randn(100,1) * 2
```

```python
#Preprocesamiento de datos
# Divide los datos en conjunto de entrenamiento y prueba
train_data, test_data, train_labels, test_labels = train_test_split(
    horas_estudio, calificaciones, test_size=0.33, random_state=42
)
```

```python

# Crea una instancia del modelo de regresión lineal
model = LinearRegression()

# Entrena el modelo con los datos de entrenamiento
model.fit(horas_estudio, calificaciones)
```

```python
# 5. Prueba del modelo
# Realiza predicciones con los datos de prueba
predictions = model.predict(test_data)
```

```python
# Graficar el modelo de regresión lineal
plt.scatter(test_data, test_labels, color='blue', label='Datos de Prueba')
plt.plot(test_data, predictions, color='red', label='Predicciones del Modelo')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificaciones')
plt.title('Modelo de Regresión Lineal')
plt.legend()
plt.show()
```

# **Entrenamiento**

```python
# Calcula el error cuadrático medio
mse = mean_squared_error(test_labels, predictions)
print(f"Error Cuadrático Medio: {mse}")
```
